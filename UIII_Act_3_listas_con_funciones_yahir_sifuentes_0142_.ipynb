```python
## Base de datos ficticia usando LISTAS
id = []
nombre = []
especialidad  = []

def agregar_estilista():
    id_est = input("Ingrese el ID del estilista: ")

    if id_est in ids:
        print("⚠️ El ID ya existe.")
    else:
        nombre = input("Nombre: ")
        especialidad = input("Especialidad (Corte, Color, Peinado): ")

        ids.append(id_est)
        nombres.append(nombre)
        especialidades.append(especialidad)

        print("✅ Estilista agregado con éxito.")

def leer_estilista():
    id_est = input("ID del estilista a consultar: ")

    if id_est in ids:
        i = ids.index(id_est)

        print(f"🔍 Resultado: [ID: {ids[i]}] "
              f"Nombre: {nombres[i]}, "
              f"Especialidad: {especialidades[i]}")
    else:
        print("❌ Estilista no encontrado.")

def listar_estilistas():
    if len(ids) == 0:
        print("📭 No hay estilistas registrados.")
    else:
        print("\n--- Lista de Estilistas ---")

        for i in range(len(ids)):
            print(f"ID: {ids[i]} | "
                  f"{nombres[i]} - "
                  f"{especialidades[i]}")

def buscar_estilista():
    nombre_buscado = input("Ingrese el nombre a buscar: ").lower()

    encontrados = False

    for i in range(len(nombres)):
        if nombre_buscado in nombres[i].lower():
            print(f"🔎 Coincidencia: ID {ids[i]} - {nombres[i]}")
            encontrados = True

    if not encontrados:
        print("🤷 No se encontraron coincidencias.")

def modificar_estilista():
    id_est = input("ID del estilista a modificar: ")

    if id_est in ids:
        i = ids.index(id_est)

        print("Deje en blanco si no desea cambiar el valor.")

        nuevo_nom = input(f"Nuevo nombre ({nombres[i]}): ")
        nueva_esp = input(f"Nueva especialidad ({especialidades[i]}): ")

        if nuevo_nom != "":
            nombres[i] = nuevo_nom

        if nueva_esp != "":
            especialidades[i] = nueva_esp

        print("📝 Datos actualizados.")

    else:
        print("❌ No existe ese ID.")

def borrar_estilista():
    id_est = input("ID del estilista a eliminar: ")

    if id_est in ids:
        i = ids.index(id_est)

        ids.pop(i)
        nombres.pop(i)
        especialidades.pop(i)

        print("🗑️ Registro eliminado.")

    else:
        print("❌ No se encontró el ID.")

# --- CICLO PRINCIPAL (Menú) ---
def menu():
    while True:
        print("\n--- 💇 SALÓN DE BELLEZA - GESTIÓN ---")
        print("1. Agregar Estilista")
        print("2. Leer (Ver uno)")
        print("3. Listar Todos")
        print("4. Buscar por Nombre")
        print("5. Modificar")
        print("6. Borrar")
        print("7. Salir")

        opcion = input("\nSeleccione una opción: ")

        if opcion == "1":
            agregar_estilista()

        elif opcion == "2":
            leer_estilista()

        elif opcion == "3":
            listar_estilistas()

        elif opcion == "4":
            buscar_estilista()

        elif opcion == "5":
            modificar_estilista()

        elif opcion == "6":
            borrar_estilista()

        elif opcion == "7":
            print("Saliendo del sistema... ¡Buen día!")
            print("Programa realizado por Yahir Sifuentes No. 0142")
            break

        else:
            print("🚫 Opción inválida, intente de nuevo.")

# Ejecutar el programa
menu()
```
