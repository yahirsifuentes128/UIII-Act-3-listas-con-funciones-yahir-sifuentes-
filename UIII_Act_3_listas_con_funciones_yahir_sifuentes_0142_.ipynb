#### lista con funciones estilista

```python
# Base de datos ficticia (Diccionario)
estilistas = {}

def agregar_estilista():
    id_est = input("Ingrese el ID del estilista: ")
    if id_est in estilistas:
        print("⚠️ El ID ya existe.")
    else:
        nombre = input("Nombre: ")
        especialidad = input("Especialidad (Corte, Color, Peinado): ")
        estilistas[id_est] = {"nombre": nombre, "especialidad": especialidad}
        print("✅ Estilista agregado con éxito.")

def leer_estilista():
    id_est = input("ID del estilista a consultar: ")
    if id_est in estilistas:
        e = estilistas[id_est]
        print(f"🔍 Resultado: [ID: {id_est}] Nombre: {e['nombre']}, Especialidad: {e['especialidad']}")
    else:
        print("❌ Estilista no encontrado.")

def listar_estilistas():
    if not estilistas:
        print("📭 No hay estilistas registrados.")
    else:
        print("\n--- Lista de Estilistas ---")
        for id_est, datos in estilistas.items():
            print(f"ID: {id_est} | {datos['nombre']} - {datos['especialidad']}")

def buscar_estilista():
    # Búsqueda por nombre (opción flexible)
    nombre_buscado = input("Ingrese el nombre a buscar: ").lower()
    encontrados = [id_e for id_e, d in estilistas.items() if nombre_buscado in d['nombre'].lower()]

    if encontrados:
        for id_e in encontrados:
            print(f"🔎 Coincidencia: ID {id_e} - {estilistas[id_e]['nombre']}")
    else:
        print("🤷 No se encontraron coincidencias.")

def modificar_estilista():
    id_est = input("ID del estilista a modificar: ")
    if id_est in estilistas:
        print("Deje en blanco si no desea cambiar el valor.")
        nuevo_nom = input(f"Nuevo nombre ({estilistas[id_est]['nombre']}): ")
        nueva_esp = input(f"Nueva especialidad ({estilistas[id_est]['especialidad']}): ")

        if nuevo_nom: estilistas[id_est]['nombre'] = nuevo_nom
        if nueva_esp: estilistas[id_est]['especialidad'] = nueva_esp
        print("📝 Datos actualizados.")
    else:
        print("❌ No existe ese ID.")

def borrar_estilista():
    id_est = input("ID del estilista a eliminar: ")
    if id_est in estilistas:
        del estilistas[id_est]
        print("🗑️ Registro eliminado.")
    else:
        print("❌ No se encontró el ID.")

# --- CICLO PRINCIPAL (Menú) ---
def menu():
    while True:
        print("\n--- 💇 SALÓN DE BELLEZA - GESTIÓN ---")
        print("1. Agregar Estilista")
        print("2. Leer (Ver uno)")
        print("3. Listar Todos")
        print("4. Buscar por Nombre")
        print("5. Modificar")
        print("6. Borrar")
        print("7. Salir")

        opcion = input("\nSeleccione una opción: ")

        if opcion == "1": agregar_estilista()
        elif opcion == "2": leer_estilista()
        elif opcion == "3": listar_estilistas()
        elif opcion == "4": buscar_estilista()
        elif opcion == "5": modificar_estilista()
        elif opcion == "6": borrar_estilista()
        elif opcion == "7":
            print("Saliendo del sistema... ¡Buen día! programa realizado por yahir sifuwntes No = 0142 .")
            break
        else:
            print("🚫 Opción inválida, intente de nuevo.")

# Ejecutar el programa
menu()
```
